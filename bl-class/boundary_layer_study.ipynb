```python
import numpy as np
from matplotlib import pyplot as plt



z_over_l = np.arange(-1, 1, .01)
phi_m_1 = 1+4.7*z_over_l
phi_m_2 = 1 + 0*z_over_l
phi_m_3 = (1 - (15*z_over_l))**(-1/4)
```

```python
fig = plt.figure()
ax = fig.gca()
ax.plot(z_over_l, phi_m_1, label='phi_m_1')
ax.plot(z_over_l, phi_m_2, label='phi_m_2')
ax.plot(z_over_l, phi_m_3, label='phi_m_3')
ax.set_yscale('log')
ax.legend()
```

```python
z_over_l_stable = np.arange(0, 1, .01)
z_over_l_neutral = 0
z_over_l_unstable = np.arange(-1, 0, .01)

phi_m_stable = 1+4.7*z_over_l_stable
phi_m_neutral = 1 + 0*z_over_l_neutral
phi_m_unstable = (1 - (15*z_over_l_unstable))**(-1/4)

fig = plt.figure()
ax = fig.gca()
ax.plot(z_over_l_stable, phi_m_stable, label='stable')
ax.plot(z_over_l_neutral, phi_m_neutral, label='neutral')
ax.plot(z_over_l_unstable, phi_m_unstable, label='unstable')
```
