```python
import xarray as xr
from datetime import datetime as dt
from os import path, getcwd, listdir
```

```python
months = [6, 7, 8, 9]
for month in months:
    lightning_root = path.join(getcwd(), '..', str(month), '6sensor_minimum')
    lightning_files = sorted(listdir(lightning_root))
    lightning_paths = [path.join(lightning_root, f) for f in lightning_files]
    for this_path in lightning_paths:
        this_lightning_ds = xr.open_dataset(this_path)
        this_lightning_date = this_lightning_ds.grid_time.values.astype('datetime64[s]').astype(dt)[0]
        big_flash_count = (this_lightning_ds.flash_event_count.data > 20).sum()
        print(f'{this_lightning_date.strftime("%Y-%m-%d")}: {big_flash_count} big flashes')
```
